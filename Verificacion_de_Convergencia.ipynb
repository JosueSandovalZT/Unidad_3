# Verificacion de Convergencia

```python
import numpy as np

# Definición de la matriz A del sistema lineal (3x3)
# Se usará para analizar la convergencia del método iterativo de Jacobi
A = np.array([[10, 1, -1],
              [2, 10, 1],
              [2, 3, 10]])

# Extracción de la matriz diagonal D de A
# np.diag(np.diag(A)) obtiene los elementos de la diagonal de A y construye una matriz diagonal con ellos
D = np.diag(np.diag(A))

# Construcción de la matriz L (parte triangular inferior negativa, sin la diagonal)
# -np.tril(A, -1) toma la parte triangular inferior de A por debajo de la diagonal principal (k=-1) y la hace negativa
L = -np.tril(A, -1)

# Construcción de la matriz U (parte triangular superior negativa, sin la diagonal)
# -np.triu(A, 1) toma la parte triangular superior de A por encima de la diagonal principal (k=1) y la hace negativa
U = -np.triu(A, 1)

# Matriz de iteración de Jacobi: T = inv(D) * (L + U)
# Equivalente a T = D^{-1} * (L+U), donde L+U es la matriz de los coeficientes no diagonales con signo cambiado
T = np.linalg.inv(D).dot(L + U)

# Cálculo de los autovalores (valores propios) de la matriz T
eigenvalues = np.linalg.eigvals(T)

# Cálculo del radio espectral: máximo valor absoluto de los autovalores
rho = max(abs(eigenvalues))

# Impresión del radio espectral
print("Radio espectral:", rho)

# Condición de convergencia del método de Jacobi:
# El método converge si y solo si el radio espectral es estrictamente menor que 1
if rho < 1:
    print("El metodo converge")
else:
    print("El metodo diverge")
```
