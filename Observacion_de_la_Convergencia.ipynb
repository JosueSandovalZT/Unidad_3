# Observacion de la Convergencia

```python
import numpy as np

# Definición de la matriz A del sistema lineal Ax = b
A = np.array([[10, 1, -1],
              [2, 10, 1],
              [2, 3, 10]])

# Definición del vector de términos independientes b
b = np.array([11, 12, 13])

# Inicialización del vector solución x con ceros (aproximación inicial)
x = np.zeros(3)

# Tolerancia para el criterio de convergencia
tol = 1e-6

# Bucle principal: máximo 50 iteraciones del método iterativo
for k in range(50):

    # Crear un nuevo vector para almacenar la solución actualizada
    x_new = np.zeros(3)

    # Actualización de la primera variable (x1)
    # Despejada de la primera ecuación: 10*x1 = 11 - x2 + x3
    x_new[0] = (11 + x[1] - x[2]) / 10

    # Actualización de la segunda variable (x2)
    # Despejada de la segunda ecuación: 10*x2 = 12 - 2*x1 - x3
    # Nota: usa x[0] (valor antiguo de x1) - característico del método de Jacobi
    x_new[1] = (12 - 2*x[0] - x[2]) / 10

    # Actualización de la tercera variable (x3)
    # Despejada de la tercera ecuación: 10*x3 = 13 - 2*x1 - 3*x2
    # Nota: usa x[0] y x[1] (valores antiguos) - método de Jacobi
    x_new[2] = (13 - 2*x[0] - 3*x[1]) / 10

    # Cálculo del error: norma euclidiana de la diferencia entre iteraciones
    error = np.linalg.norm(x_new - x)

    # Mostrar información de la iteración actual
    print("Iteracion:", k+1, "Solucion:", x_new, "Error:", error)

    # Criterio de parada: si el error es menor que la tolerancia, terminar
    if error < tol:
        break

    # Actualizar x para la siguiente iteración
    x = x_new
```
