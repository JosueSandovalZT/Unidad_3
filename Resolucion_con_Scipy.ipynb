# Resolucion con Scipy

```python
import numpy as np
from scipy.optimize import fsolve

# Definición del sistema de ecuaciones no lineales
# La función recibe un vector de variables y devuelve los valores de las ecuaciones
def sistema(vars):
    x, y = vars  # Desempaqueta las dos variables del sistema

    # Primera ecuación: circunferencia de radio 2 centrada en el origen
    f1 = x**2 + y**2 - 4

    # Segunda ecuación: recta con pendiente 1 que pasa por (0, -1) y (1, 0)
    f2 = x - y - 1

    return [f1, f2]  # Retorna ambas ecuaciones evaluadas

# Resolución numérica del sistema no lineal
# fsolve busca la raíz del sistema (valores de x,y que hacen f1=0 y f2=0)
# Parámetros: función del sistema, aproximación inicial [1,1]
sol = fsolve(sistema, [1,1])

# Muestra la solución aproximada encontrada
print("Solucion aproximada", sol)
```
