# Método de Gauss - Seidel

```python
import numpy as np   # Biblioteca para cálculos numéricos

x = np.zeros(3)   # Vector inicial de aproximación para las variables

tol = 1e-6   # Tolerancia permitida para el error

for k in range(100):   # Bucle de iteraciones

    x_old = x.copy()   # Guarda la solución anterior para calcular el error

    x[0] = (11 - x[1] + x[2]) / 10   # Actualización de la variable x usando

    x[1] = (12 - 2 * x[0] - x[2]) / 10   # Actualización de y usando el nuevo

    x[2] = (13 - 2 * x[0] - 3 * x[1]) / 10   # Actualización de z usando valores

    if np.linalg.norm(x - x_old) < tol:   # Verifica si el cambio entre iter
        break   # Detiene el proceso si se alcanza la convergencia

print("solucion Gauss-seidel:", x)
print("iteraciones:", k + 1)
```
