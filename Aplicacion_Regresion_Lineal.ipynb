# Aplicacion Regresion Lineal

```python
import numpy as np

# Matriz de diseño X para un modelo de regresión lineal simple con intercepto
# Primera columna: unos (término independiente o intercepto)
# Segunda columna: valores de la variable predictora (x)
# Cada fila corresponde a una observación
X = np.array([[1, 1],
              [1, 2],
              [1, 3],
              [1, 4]])

# Vector de respuestas y (variable dependiente)
# Valores observados correspondientes a cada x
y = np.array([6, 5, 7, 10])

# Transposición de la matriz de diseño
# Se calcula X^T para poder realizar el producto matricial necesario
XT = X.T

# Cálculo de los coeficientes del modelo de regresión lineal por mínimos cuadrados ordinarios (MCO)
# Fórmula: β = (X^T · X)^{-1} · X^T · y
# Paso a paso:
#   1. XT.dot(X)  -> X^T · X  (matriz de momentos)
#   2. np.linalg.inv(...) -> inversa de (X^T · X)
#   3. .dot(XT) -> multiplicar por X^T
#   4. .dot(y) -> multiplicar por el vector y
beta = np.linalg.inv(XT.dot(X)).dot(XT).dot(y)

# Impresión de los coeficientes estimados:
#   beta[0] = intercepto (ordenada al origen)
#   beta[1] = pendiente (efecto de x sobre y)
print("Coeficientes del modelo:", beta)
```
