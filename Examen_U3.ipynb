# Parte B - Implementación Numérica

```python
import numpy as np

A = np.array([[10, 2, 1],   # Primera ecuación
              [2, 10, 3],   # Segunda ecuación
              [1, 3, 10]])  # Tercera ecuación

# Vector de términos independientes
b = np.array([9, 13, 14])

# Vector inicial de aproximación (x0, y0, z0)
x = np.zeros(3)

# Tolerancia para el criterio de paro
tol = 1e-6

# Número máximo de iteraciones permitidas
max_iter = 100

# Bucle principal de iteraciones
for k in range(max_iter):
    x_new = np.zeros(3)

    x_new[0] = (9 - x[1] + x[2]) / 10
    x_new[1] = (13 - 2 * x[0] - x[2]) / 10
    x_new[2] = (14 - 2 * x[0] - x[1]) / 10

    error = np.linalg.norm(x_new - x, np.inf)

    if np.linalg.norm(x_new - x) < tol:
        x = x_new
        break
    x = x_new

# Mostrar resultados finales
print("Número de iteraciones:", k + 1)
print("Error final:", error)
print("Solución final:", x_new)
```

# Parte C - Análisis de convergencia

```python
import numpy as np # Importa la biblioteca NumPy para realizar cálculos numéricos y aproximación

A = np.array([[10, 2, 1],   # Primera ecuación
              [2, 10, 3],   # Segunda ecuación
              [1, 3, 10]])  # Tercera ecuación

D = np.diag(np.diag(A)) # Extrae la matriz diagonal de A.

L = -np.tril(A,-1) # Obtiene la parte triangular inferior de la matriz A con signo negativo

U = -np.triu(A,1) # Obtiene la parte triangular superior de la matriz A con signo negativo

T = np.linalg.inv(D).dot(L+U)

eigenvalues = np.linalg.eigvals(T)

rho = max(abs(eigenvalues))

print("Eigenvalores:", eigenvalues)
print("Radio espectral:", rho)

if rho < 1:
    print("El metodo converge")
else:
    print("El método diverge")
```
