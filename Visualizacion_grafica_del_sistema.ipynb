# Ejercicio 1: Visualización gráfica del sistema

```python
import numpy as np # Importa la biblioteca NumPy para cálculos númericos
import matplotlib.pyplot as plt # Importa MatplotLib para generar graficas

X = np.linspace(-3, 3, 400)

y1 = np.sqrt(4 - X ** 2)
y2 = -np.sqrt(4 - X ** 2)

y_line = X - 1

plt.plot(X, y1) # Grafica la mitad superior del circulo
plt.plot(X, y2) # Grafica la mitad inferior del circulo
plt.plot(X, y_line) # Grafica la recta del sistema

plt.xlabel("X") # Etiqueta del eje horizontal
plt.ylabel("y") # Etiqueta del eje vertical
plt.title("Sistema de ecuaciones no lineales") # Título del gráfico
plt.grid() # Activa la cuadrícula para facilitar la visualización

plt.show()
```
