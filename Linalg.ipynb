# Linalg

```python
import numpy as np

# Definición de la matriz A del sistema lineal Ax = b
A = np.array([[10, 1, -1],
             [2, 10, 1],
             [2, 3, 10]])

# Definición del vector de términos independientes b
b = np.array([11, 12, 13])

# Resolución directa del sistema lineal mediante descomposición LU (método de eliminación de Gauss)
# np.linalg.solve calcula la solución exacta (con precisión numérica) del sistema Ax = b
x = np.linalg.solve(A, b)

# Impresión de la solución del sistema
print("Solucion del siistema:", x)  # Nota: hay un pequeño error tipográfico: "siistema" en lugar de "sistema"
```
