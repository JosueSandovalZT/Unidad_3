# Evaluacion del Error

```python
import numpy as np

# Valores aproximados de la solución del sistema no lineal:
#   x^2 + y^2 = 4
#   x - y = 1
# Estos valores provienen de la solución aproximada obtenida con fsolve
x = 1.822
y = 0.822

# Evaluación de la primera ecuación en la solución aproximada
# Se espera que el resultado sea cercano a 0 si la aproximación es buena
f1 = x**2 + y**2 - 4

# Evaluación de la segunda ecuación en la solución aproximada
# También debe ser cercano a 0
f2 = x - y - 1

# Impresión del residuo (error) de cada ecuación
# Muestra qué tan cerca está la solución aproximada de satisfacer cada ecuación
print("Error ecuacion 1:", f1)
print("Error ecuacion 2:", f2)
```
